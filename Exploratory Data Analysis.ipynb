```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data_set=pd.read_csv("/Users/ranitaghosh/Desktop/SampleSuperstore.csv")
data_set
```

```python
data_set.info()
```

```python
data_set.describe()
```

```python
sns.set_style("whitegrid")
sns.pairplot(data_set,hue="Region",palette='bwr')
```

```python
data_set['Ship Mode'].unique()
```

```python
segment=data_set.Segment.value_counts().reset_index()
segment.columns=("Segment","Count")
segment
```

```python
plt.pie(x="Count",labels="Segment",data=segment,radius=3,autopct="%.2f",pctdistance=0.4)
```

```python
plt.figure(figsize=(9,5))
plt.title('CATEGORIES VS REGION')
sns.countplot(x=data_set['Category'],hue=data_set['Region'],palette='rocket')
plt.xticks()
```

```python
high_sale=data_set[data_set["Sales"]>3000]
high_sale.head()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=high_sale["Sales"],y=high_sale["State"],data=data_set)
```

```python
data_set.Sales.max()
```

```python
data_set.Sales.min()
```

```python
plt.figure(figsize=(9,5))
low_sale=data_set[data_set["Sales"]<1]
sns.barplot(y=low_sale["State"],x=low_sale["Sales"],data=low_sale,palette='rocket')
```

```python
low_profit=data_set[data_set["Profit"]<1]
low_profit.head()
```

```python
data=['Sales','Quantity','Discount','Profit']
plt.figure(figsize=(9,5))
sns.heatmap(data_set[data].corr(),annot=True )
```
